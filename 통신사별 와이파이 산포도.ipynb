```python
skx=[]
sky=[]
lgx = []
lgy = []
ktx = []
kty = []    #각각의 통신사의 와이파이 좌표 리스트
import csv
f= open('전국무료와이파이표준데이터.csv')
data=csv.reader(f)
next(data)
for row in data:
    if 'SK' in row[5] :
        skx.append(float(row[13]))    #통신사가 SK인 무료 와이파이의 위도와 경도
        sky.append(float(row[12]))
        
    elif 'LG' in row[5] :
        lgx.append(float(row[13]))    #통신사가 LG인 무료 와이파이의 위도와 경도
        lgy.append(float(row[12]))
    
    elif 'KT' in row[5] :
        ktx.append(float(row[13]))    #통신사가 KT인 무료 와이파이의 위도와 경도
        kty.append(float(row[12]))

import matplotlib.pyplot as plt
plt.figure(figsize = (10,10), dpi = 300) 
plt.rc('font',family='Malgun Gothic')    #글씨 폰트

plt.title('전국 통신사별 무료 와이파이')    #제목

#plt.scatter(skx, sky, color = 'darkblue', alpha=0.3, label = 'SK')    #SK 통신사의 무료 와이파이에 대한 산포도
#plt.scatter(lgx, lgy, color = 'darkcyan', alpha=0.3, label = 'LG')    #LG 통신사의 무료 와이파이에 대한 산포도
#plt.scatter(ktx, kty, color = 'cyan', alpha=0.3, label = 'KT')    #KT 통신사의 무료 와이파이에 대한 산포도
plt.legend()

plt.xlim(124,132)
plt.ylim(33,40)    #우리나라(남한)의 경도의 위도범위
plt.xlabel('경도')
plt.ylabel('위도')
#plt.savefig('산포도 sk.png')
plt.show()
```
