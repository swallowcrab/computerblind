```python
skx=[]
sky=[]
lgx = []
lgy = []
ktx = []
kty = []    #    
import csv
f= open('.csv')
data=csv.reader(f)
next(data)
for row in data:
    if 'SK' in row[5] :
        skx.append(float(row[13]))    # SK    
        sky.append(float(row[12]))
        
    elif 'LG' in row[5] :
        lgx.append(float(row[13]))    # LG    
        lgy.append(float(row[12]))
    
    elif 'KT' in row[5] :
        ktx.append(float(row[13]))    # KT    
        kty.append(float(row[12]))

import matplotlib.pyplot as plt
plt.figure(figsize = (10,10), dpi = 300) 
plt.rc('font',family='Malgun Gothic')    # 

plt.title('   ')    #

#plt.scatter(skx, sky, color = 'darkblue', alpha=0.3, label = 'SK')    #SK     
#plt.scatter(lgx, lgy, color = 'darkcyan', alpha=0.3, label = 'LG')    #LG     
#plt.scatter(ktx, kty, color = 'cyan', alpha=0.3, label = 'KT')    #KT     
plt.legend()

plt.xlim(124,132)
plt.ylim(33,40)    #()  
plt.xlabel('')
plt.ylabel('')
#plt.savefig(' sk.png')
plt.show()
```
