```python
space=[]
wifisp=[]
wificn=[]
sd={}
k=[]
m=[]
n=[]   


num=int(input('와이파이가 가장 많이 설치된 지역 순위와 비율; 몇 위까지 알고 싶나요?'))  


import csv
f= open('전국무료와이파이표준데이터.csv')    


data=csv.reader(f)
next(data)
for row in data:
    space.append(row[3])   


exsp=space


for lst in space:
    try: sd[lst] += 1
    except: sd[lst]=1    

        
import operator

so = sorted(sd.items(), key=operator.itemgetter(1), reverse=True)    

for i, j in so :
    wifisp.append(i)    
    wificn.append(j)    

for i in range(num-1):    

    k.append(wifisp[i])
    m.append(wificn[i])    


for i in range(len(k)):
    n.append((str(k[i])+': '+str(m[i])))    

print(n)    

import matplotlib.pyplot as plt
plt.rc('font',family='Malgun Gothic', size=5)
plt.figure(figsize = (5,3), dpi = 300)
plt.pie(m, labels=k, autopct='%1.1f%%', startangle=90)
plt.axis('equal')
#plt.savefig('설치된 개수 num개.png')
plt.show()
```
