```python
kt=[]
lg=[]
sk=[]
gt=[]    #  

import csv
f= open('.csv')
data=csv.reader(f)
next(data)
for row in data:
    if 'KT' in row[5] :
        kt.append(row[5])    #KT    
        
    elif 'LG' in row[5] :
        lg.append(row[5])    #LG    
    
    elif 'SK' in row[5] :
        sk.append(row[5])    #SK    
        
    #else:
        #gt.append(row[5])    #    
        
import matplotlib.pyplot as plt
plt.rc('font',family='Malgun Gothic')
plt.figure(figsize=(6,4), dpi=300)
explode=(0.1,0,0)    #KT  
labels= 'KT', 'LG', 'SK'
colors= ['darkcyan', 'cyan', 'lightcyan']
plt.pie([len(kt),len(lg), len(sk)],explode=explode,labels=labels,colors=colors,autopct='%1.3f%%', startangle=90)
    #

plt.axis('equal')
plt.savefig('data3up.png') 
plt.show()
```
