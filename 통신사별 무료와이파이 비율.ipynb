```python
kt=[]
lg=[]
sk=[]
gt=[]    #각각의 통신사의 리스트

import csv
f= open('전국무료와이파이표준데이터.csv')
data=csv.reader(f)
next(data)
for row in data:
    if 'KT' in row[5] :
        kt.append(row[5])    #KT 통신사의 무료 와이파이 개수
        
    elif 'LG' in row[5] :
        lg.append(row[5])    #LG 통신사의 무료 와이파이 개수
    
    elif 'SK' in row[5] :
        sk.append(row[5])    #SK 통신사의 무료 와이파이 개수
        
    #else:
        #gt.append(row[5])    #기타 통신사의 무료 와이파이 개수
        
import matplotlib.pyplot as plt
plt.rc('font',family='Malgun Gothic')
plt.figure(figsize=(6,4), dpi=300)
explode=(0.1,0,0)    #KT통신사만 따로 떼어놓음
labels= 'KT', 'LG', 'SK'
colors= ['darkcyan', 'cyan', 'lightcyan']
plt.pie([len(kt),len(lg), len(sk)],explode=explode,labels=labels,colors=colors,autopct='%1.3f%%', startangle=90)
    #원그래프

plt.axis('equal')
plt.savefig('data3up.png') 
plt.show()
```
